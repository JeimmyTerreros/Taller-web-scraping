```python
import bs4 as bs 
import urllib.request
import re
```

```python
animales = ["https://es.wikipedia.org/wiki/Canis_familiaris",
"https://es.wikipedia.org/wiki/Oryctolagus_cuniculus",
"https://es.wikipedia.org/wiki/Panthera_leo",
"https://es.wikipedia.org/wiki/Psittacoidea",
"https://es.wikipedia.org/wiki/Panthera_tigris",
"https://es.wikipedia.org/wiki/Dinosauria",
"https://es.wikipedia.org/wiki/Felis_silvestris_catus",
"https://es.wikipedia.org/wiki/Delphinidae",
"https://es.wikipedia.org/wiki/Rhopalocera",
"https://es.wikipedia.org/wiki/Panthera_onca"]
```

```python
todossources = []
for animal in animales:
    request = urllib.request.Request(animal, headers={'User-Agent': 'Mozilla/5.0'})
    webpage = urllib.request.urlopen(request)
    source = webpage.read()
    webpage.close()
    todossources.append(source)
   
```

```python
for animal in animales:
    soup = bs.BeautifulSoup(source, "html.parser")
    print(soup.find("h1").text)
```
